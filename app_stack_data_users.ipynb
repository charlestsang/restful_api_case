```python
import requests
import json
import os
import pandas as pd
```

```python
#stack questions
#link = "https://api.stackexchange.com/2.3/questions?order=desc&sort=activity&site=stackoverflow"

#stack answers
#link = 'https://api.stackexchange.com/2.3/answers?order=desc&sort=activity&site=stackoverflow'

#stack users
link = 'https://api.stackexchange.com/2.3/users?order=desc&sort=reputation&site=stackoverflow'

response = requests.get(link)

re = response.json()
re
```

```python
re.keys()
```

```python
re['items']
```

```python
type(re['items'])
```

```python

for data in re['items']:
    if data['reputation'] >= 10:
        print(data['user_id'])
        print(data['user_type'])
        print(data['display_name'])
        print(data['is_employee'])
        print(data['account_id'])
        print(data['reputation'])
        print(data['creation_date'])
        #print(data['location'])
    else:
        print("skipped")
    print()
```

```python
df = pd.DataFrame(re['items'])
df.head()
```

```python
#col =['account_id','user_id','user_type','display_name',
#      'is_employee','location','reputation','accept_rate','creation_date']

#select sample cols
col = ['user_id','user_type','display_name','location']


data = df[col]
data.head()
```

```python
# extract data to csv file

outname = 'stack_users_sample.csv'
outdir = './data'

if not os.path.exists(outdir):
    os.mkdir(outdir)

fullname = os.path.join(outdir, outname)    

data.to_csv(fullname, sep='\t', encoding='utf-8')
```
