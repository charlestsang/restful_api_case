```python
import requests
import json
import os
import pandas as pd
```

```python
# stack questions
link = "https://api.stackexchange.com/2.3/questions?order=desc&sort=activity&site=stackoverflow"
response = requests.get(link)

re = response.json()
re
```

```python
re.keys()
```

```python
re['items']
```

```python
type(re['items'])
```

```python

for data in re['items']:
    if data['answer_count'] <= 10:
        print(data['owner']['user_id'])
        print(data['owner']['user_type'])
        print(data['owner']['display_name'])
        print(data['title'])
        print(data['link'])
        print(data['question_id'])
        print(data['answer_count'])
        print(data['score'])
        print(data['creation_date'])
    else:
        print("skipped")
    print()
```

```python
df = pd.DataFrame(re['items'])
df.head()
```

```python
df["owner"][0]
```

```python
sample_type = type(df["owner"][0])
print(f"The data type before using apply function: {sample_type}\n")
```

```python
df_owner = pd.json_normalize(df["owner"],max_level=1)
df_owner =df_owner.drop('link',axis=1)
```

```python
df_owner.head()
```

```python
result = pd.concat([df, df_owner], axis =1)
result.head()
```

```python
#col =['account_id','user_id','user_type','display_name','title','link',
#      'question_id','is_answered','view_count','answer_count','score','creation_date']
#
# selected sample col
col = ['user_id','question_id','title']

data = result[col]
data.head()
```

```python
# extract data to csv file

outname = 'stack_questions_sample.csv'
outdir = './data'

if not os.path.exists(outdir):
    os.mkdir(outdir)

fullname = os.path.join(outdir, outname)    

data.to_csv(fullname, sep='\t', encoding='utf-8')
```
